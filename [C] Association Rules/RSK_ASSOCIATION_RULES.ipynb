```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder 
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
df = pd.get_dummies(titanic)
```

```python
len(df.columns)
```

```python
df = pd.DataFrame(df, columns=df.columns)
```

```python
frequent_itemsets = apriori(df.astype('bool'), min_support=0.1, use_colnames=True)
```

```python
pd.concat([frequent_itemsets.head(), frequent_itemsets.tail()], axis=0)
```

```python
rules = association_rules(frequent_itemsets, metric='lift',min_threshold=0.7)
rules
```

```python
new = rules.sort_values('lift', ascending=False)
```

```python
new[new['lift'] > 1]
```
