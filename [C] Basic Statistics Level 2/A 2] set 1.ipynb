```python
import matplotlib.pyplot as plt 
from scipy.stats import binom
import seaborn as sns
import pandas as pd
import numpy as np
```

# Q 1] Plot the data, find the outliers and find out  μ,σ,σ^2

```python
comp = ['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers','Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways','Warner-Lambert']
measureX = [24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00]
measureX = np.array(measureX)
```

```python
# mean, standard deviation, variance 
pd.DataFrame([['Mean',measureX.mean()],
              ['Standard dev',measureX.std()],
              ['Variance',measureX.var()]],
             columns=['Moment','Values'])
```

```python
# barplot 
sns.barplot(x=comp,
            y=measureX)
plt.xticks(rotation=90)
plt.xlabel('Company Name')
plt.ylabel('Measure X')
plt.show()
```

```python
# violinplot
# majority of data is centered in range 20 to 40
sns.violinplot(data=measureX,
               palette='Set3')
plt.show()
```

```python
# boxplot
# outlier is present 
point = plt.boxplot(measureX)
plt.show()
```

```python
# only 1 outlier is present 
print(f"No. of Outliers present : {len(point['fliers'])}")
print(f"Outlier datapoint : {point['fliers'][0].get_data()[1][0]}")
```

## Q 4] What is the probability that at least one in five attempted telephone calls reaches the wrong number

```python
# number of trials 
n = 5
# probability of getting a wrong number in each call
p = 1/200
prob = binom.pmf(1,n=n,p=p)*100
print(f'Probability that atleast one in five attempted telephone calls reaches wrong number is {round(prob,2)}%')
```

# Q 5] ii] Is the venture likely to be successful? Explain

```python
x = np.array([-2000,-1000,0,1000,2000,3000])
pX = np.array([0.1,0.1,0.2,0.2,0.3,0.1])
expectReturn = (x*pX).sum()
print(f"monetary outcome of bussiness is {expectReturn}") 
```

### since monetary outcome is positive, it can be concluded that venture is likely to be successful
